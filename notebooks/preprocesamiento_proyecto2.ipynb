El proceso de conversion a tensores en este caso es fácil, tan solo se debe usar ToTensor de torchvision y eso ya nos dejará las imágenes como tensores normalizados. También para reescalarlas a 128x128 usamos el transform de Resize.

## Dataset de training y validation

Primero trabajemos con el dataset de training ya que este solo cuenta con ejemplos "good", el proceso de lectura de los datasets de trainining será un tanto distinto

```python
import torch
from PIL import Image
import torchvision.transforms as transforms

transform = transforms.Compose([
    transforms.Resize((128, 128)), # Esto pasa las imágenes de la resu original a la que pide el profe
    transforms.ToTensor() # ToTensor para que si normalice
])


def load_training_dataset(dir):
    result = []

    for file in os.listdir(dir):
        path = os.path.join(dir, file)
        img = Image.open(path).convert("RGB")

        tensor = transform(img)
        result.append(tensor)

    return result
```

```python
train_cable_list = load_training_dataset(train_cable_dir)
train_capsule_list = load_training_dataset(train_capsule_dir)
train_carpet_list = load_training_dataset(train_carpet_dir)
train_grid_list = load_training_dataset(train_grid_dir)
train_hazelnut_list = load_training_dataset(train_hazelnut_dir)
train_leather_list = load_training_dataset(train_leather_dir)
train_metal_nut_list = load_training_dataset(train_metal_nut_dir)
train_screw_list = load_training_dataset(train_screw_dir)
train_transistor_list = load_training_dataset(train_transistor_dir)
train_wood_list = load_training_dataset(train_wood_dir)
```

```python
train_cable_tensor = torch.stack(train_cable_list)
train_capsule_tensor = torch.stack(train_capsule_list)
train_carpet_tensor = torch.stack(train_carpet_list)
train_grid_tensor = torch.stack(train_grid_list)
train_hazelnut_tensor = torch.stack(train_hazelnut_list)
train_leather_tensor = torch.stack(train_leather_list)
train_metal_nut_tensor = torch.stack(train_metal_nut_list)
train_screw_tensor = torch.stack(train_screw_list)
train_transistor_tensor = torch.stack(train_transistor_list)
train_wood_tensor = torch.stack(train_wood_list)
```

```python
print(train_cable_tensor.shape)
print(train_capsule_tensor.shape)
print(train_carpet_tensor.shape)
print(train_grid_tensor.shape)
print(train_hazelnut_tensor.shape)
print(train_leather_tensor.shape)
print(train_metal_nut_tensor.shape)
print(train_screw_tensor.shape)
print(train_transistor_tensor.shape)
print(train_wood_tensor.shape)
```

```python
def split_train_val(X, val_ratio=0.15, seed=42):
    n = len(X)

    generator = torch.Generator().manual_seed(seed)
    indices = torch.randperm(n, generator=generator)

    n_val = int(n * val_ratio)

    val_indices = indices[:n_val]
    train_indices = indices[n_val:]

    X_train = X[train_indices]
    X_val = X[val_indices]

    return X_train, X_val
```

```python
# Ya acaá ahora si queda dividido en training y validation
cable_train, cable_val = split_train_val(train_cable_tensor, val_ratio=0.15)
capsule_train, capsule_val = split_train_val(train_capsule_tensor, val_ratio=0.15)
carpet_train, carpet_val = split_train_val(train_carpet_tensor, val_ratio=0.15)
grid_train, grid_val = split_train_val(train_grid_tensor, val_ratio=0.15)
hazelnut_train, hazelnut_val = split_train_val(train_hazelnut_tensor, val_ratio=0.15)
leather_train, leather_val = split_train_val(train_leather_tensor, val_ratio=0.15)
metal_nut_train, metal_nut_val = split_train_val(train_metal_nut_tensor, val_ratio=0.15)
screw_train, screw_val = split_train_val(train_screw_tensor, val_ratio=0.15)
transistor_train, transistor_val = split_train_val(train_transistor_tensor, val_ratio=0.15)
wood_train, wood_val = split_train_val(train_wood_tensor, val_ratio=0.15)
```

```python
X_train = torch.cat([
    cable_train,
    capsule_train,
    carpet_train,
    grid_train,
    hazelnut_train,
    leather_train,
    metal_nut_train,
    screw_train,
    transistor_train,
    wood_train
], dim=0)

X_val = torch.cat([
    cable_val,
    capsule_val,
    carpet_val,
    grid_val,
    hazelnut_val,
    leather_val,
    metal_nut_val,
    screw_val,
    transistor_val,
    wood_val
], dim=0)
```

```python
print("Train:", X_train.shape)
print("Val:", X_val.shape)
```

```python
# También creemos tensores "Label", no se usarán para entrenar, son solo metadata para que luego los gráficos muestren quien es cada clase

CLASS_TO_IDX = {
    "cable": 0,
    "capsule": 1,
    "carpet": 2,
    "grid": 3,
    "hazelnut": 4,
    "leather": 5,
    "metal_nut": 6,
    "screw": 7,
    "transistor": 8,
    "wood": 9
}

y_train = torch.cat([
    torch.full((len(cable_train),), CLASS_TO_IDX["cable"]),
    torch.full((len(capsule_train),), CLASS_TO_IDX["capsule"]),
    torch.full((len(carpet_train),), CLASS_TO_IDX["carpet"]),
    torch.full((len(grid_train),), CLASS_TO_IDX["grid"]),
    torch.full((len(hazelnut_train),), CLASS_TO_IDX["hazelnut"]),
    torch.full((len(leather_train),), CLASS_TO_IDX["leather"]),
    torch.full((len(metal_nut_train),), CLASS_TO_IDX["metal_nut"]),
    torch.full((len(screw_train),), CLASS_TO_IDX["screw"]),
    torch.full((len(transistor_train),), CLASS_TO_IDX["transistor"]),
    torch.full((len(wood_train),), CLASS_TO_IDX["wood"])
])

y_val = torch.cat([
    torch.full((len(cable_val),), CLASS_TO_IDX["cable"]),
    torch.full((len(capsule_val),), CLASS_TO_IDX["capsule"]),
    torch.full((len(carpet_val),), CLASS_TO_IDX["carpet"]),
    torch.full((len(grid_val),), CLASS_TO_IDX["grid"]),
    torch.full((len(hazelnut_val),), CLASS_TO_IDX["hazelnut"]),
    torch.full((len(leather_val),), CLASS_TO_IDX["leather"]),
    torch.full((len(metal_nut_val),), CLASS_TO_IDX["metal_nut"]),
    torch.full((len(screw_val),), CLASS_TO_IDX["screw"]),
    torch.full((len(transistor_val),), CLASS_TO_IDX["transistor"]),
    torch.full((len(wood_val),), CLASS_TO_IDX["wood"])
])
```

## Dataset de testing

Este se debe manejar un tanto distinto debido a que en este si se tienen clases de defectos. Por lo que se seguirá la misma idea anterior de generar un tensor con todos los tensores correspondientes a la imágen de Testing, y también la creación de un tensor Y con metadata. Solo que en este caso el tensor de metadata no solo contendrá la clase de la imágen, si no que también el tipo de defecto. Cabe recalcar que los tipos de defecto varían en cada clase, por lo que el encoding usado para representarlos dependerá de la clase.  

```python
data_dir = Path(dataset_path)

test_cable_dir = data_dir / "cable" / "test"
test_capsule_dir = data_dir / "capsule" / "test"
test_carpet_dir = data_dir / "carpet" / "test"
test_grid_dir = data_dir / "grid" / "test"
test_hazelnut_dir = data_dir / "hazelnut" / "test"
test_leather_dir = data_dir / "leather" / "test"
test_metal_nut_dir = data_dir / "metal_nut" / "test"
test_screw_dir = data_dir / "screw" / "test"
test_transistor_dir = data_dir / "transistor" / "test"
test_wood_dir = data_dir / "wood" / "test"
```

```python
def load_testing_dataset(dir, object_class):

    X_result = []
    y_result = []
    defect_mapping = {}

    # El index 0 está reservado para la clase good
    i = 1
    for defect_dir in dir.iterdir():

        if not defect_dir.is_dir():
            continue

        defect_type = defect_dir.name

        defect_code = 0 if defect_type == "good" else i

        defect_mapping[defect_type] = defect_code

        for file in sorted(os.listdir(defect_dir)): # Changed 'dir' to 'defect_dir'
            path = os.path.join(defect_dir, file)
            img = Image.open(path).convert("RGB")

            img_tensor = transform(img)

            X_result.append(img_tensor)
            y_result.append(torch.tensor([object_class, defect_code]))

        # No sumamos en la iteracion de good
        i = i if defect_code == 0 else i+1

    return X_result, y_result, defect_mapping
```

```python
test_cable_list,      y_cable_list,      cable_mapping      = load_testing_dataset(test_cable_dir, CLASS_TO_IDX["cable"])
test_capsule_list,    y_capsule_list,    capsule_mapping    = load_testing_dataset(test_capsule_dir, CLASS_TO_IDX["capsule"])
test_carpet_list,     y_carpet_list,     carpet_mapping     = load_testing_dataset(test_carpet_dir, CLASS_TO_IDX["carpet"])
test_grid_list,       y_grid_list,       grid_mapping       = load_testing_dataset(test_grid_dir, CLASS_TO_IDX["grid"])
test_hazelnut_list,   y_hazelnut_list,   hazelnut_mapping   = load_testing_dataset(test_hazelnut_dir, CLASS_TO_IDX["hazelnut"])
test_leather_list,    y_leather_list,    leather_mapping    = load_testing_dataset(test_leather_dir, CLASS_TO_IDX["leather"])
test_metal_nut_list,  y_metal_nut_list,  metal_nut_mapping  = load_testing_dataset(test_metal_nut_dir, CLASS_TO_IDX["metal_nut"])
test_screw_list,      y_screw_list,      screw_mapping      = load_testing_dataset(test_screw_dir, CLASS_TO_IDX["screw"])
test_transistor_list, y_transistor_list, transistor_mapping = load_testing_dataset(test_transistor_dir, CLASS_TO_IDX["transistor"])
test_wood_list,       y_wood_list,       wood_mapping       = load_testing_dataset(test_wood_dir, CLASS_TO_IDX["wood"])
```

```python
test_cable_tensor = torch.stack(test_cable_list)
test_capsule_tensor = torch.stack(test_capsule_list)
test_carpet_tensor = torch.stack(test_carpet_list)
test_grid_tensor = torch.stack(test_grid_list)
test_hazelnut_tensor = torch.stack(test_hazelnut_list)
test_leather_tensor = torch.stack(test_leather_list)
test_metal_nut_tensor = torch.stack(test_metal_nut_list)
test_screw_tensor = torch.stack(test_screw_list)
test_transistor_tensor = torch.stack(test_transistor_list)
test_wood_tensor = torch.stack(test_wood_list)

y_cable_tensor = torch.stack(y_cable_list)
y_capsule_tensor = torch.stack(y_capsule_list)
y_carpet_tensor = torch.stack(y_carpet_list)
y_grid_tensor = torch.stack(y_grid_list)
y_hazelnut_tensor = torch.stack(y_hazelnut_list)
y_leather_tensor = torch.stack(y_leather_list)
y_metal_nut_tensor = torch.stack(y_metal_nut_list)
y_screw_tensor = torch.stack(y_screw_list)
y_transistor_tensor = torch.stack(y_transistor_list)
y_wood_tensor = torch.stack(y_wood_list)
```

## Serializar

Finalmente serializamos los tensores en un .pt para no tener que volver a correr el script
